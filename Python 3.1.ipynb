# Mission 5

Create a function that returns the sum of two numbers when executed.  
Also, your function must return a sentence that simply indicates if the sum of the two numbers chosen by the user is:
* less than 10
* between 10 and 50
* greater than 50  
Again, you must use `return`.

```python
def mission5(a,b):
  if (a+b) <10 :
    return ("less than 10", a+b)
  elif (a+b) >= 10 and (a+b) <= 50 :
    return ("between 10 and 50", a+b)
  elif (a+b) > 50 :
    return ("greater than 50", a+b)
mission5(20,30)
```

## Exercice 5 : Trouver les nombres premiers

Ã‰crire un programme qui trouve et affiche tous les nombres premiers entre 1 et 100.

```python
def est_prime(n):
  if n<=1:
    return False
  for i in range(2,int(n**0.5) +1):
    if n % i == 0:
      return False
  return True


for i in range(1,101):
  if est_prime(i):
    print(i)
```

## Exercice 7 : Calcul du PGCD

Ã‰crire une fonction qui calcule le Plus Grand Commun Diviseur (PGCD) de deux nombres donnÃ©s par l'utilisateur en utilisant l'algorithme d'Euclide.

```python
# Solution ici :
def pgcd(a, b):
    while b:
        a, b = b, a % b
    return a

n1 = int(input("Entrez le premier nombre : "))
n2 = int(input("Entrez le deuxiÃ¨me nombre : "))
print(f"Le PGCD de {n1} et {n2} est {pgcd(n1, n2)}.")
```

## Exercice 9 : Recherche dichotomique

ImplÃ©menter une recherche dichotomique pour trouver un nombre dans une liste triÃ©e. Retourner l'indice si trouvÃ©, sinon -1.

```python
# Solution ici :
def recherche_dichotomique(liste, cible):
    gauche, droite = 0, len(liste) - 1
    while gauche <= droite:
        milieu = (gauche + droite) // 2
        if liste[milieu] == cible:
            return milieu
        elif liste[milieu] < cible:
            gauche = milieu + 1
        else:
            droite = milieu - 1
    return -1

liste = [1, 3, 5, 7, 9, 11]
cible = int(input("Entrez un nombre Ã  rechercher : "))
resultat = recherche_dichotomique(liste, cible)
if resultat != -1:
    print(f"TrouvÃ© Ã  l'indice {resultat}.")
else:
    print("Non trouvÃ©.")
```

## Exercice 10 : Sudoku Validator

Ã‰crire une fonction qui vÃ©rifie si une grille 9x9 est une solution valide pour un Sudoku.

```python
# Solution ici :
def est_valide(grille):
    def verifie_bloc(bloc):
        return sorted(bloc) == list(range(1, 10))

    for ligne in grille:
        if not verifie_bloc(ligne):
            return False

    for col in zip(*grille):
        if not verifie_bloc(col):
            return False

    for i in range(0, 9, 3):
        for j in range(0, 9, 3):
            bloc = [grille[x][y] for x in range(i, i + 3) for y in range(j, j + 3)]
            if not verifie_bloc(bloc):
                return False

    return True

grille = [
    [5, 3, 4, 6, 7, 8, 9, 1, 2],
    [6, 7, 2, 1, 9, 5, 3, 4, 8],
    [1, 9, 8, 3, 4, 2, 5, 6, 7],
    [8, 5, 9, 7, 6, 1, 4, 2, 3],
    [4, 2, 6, 8, 5, 3, 7, 9, 1],
    [7, 1, 3, 9, 2, 4, 8, 5, 6],
    [9, 6, 1, 5, 3, 7, 2, 8, 4],
    [2, 8, 7, 4, 1, 9, 6, 3, 5],
    [3, 4, 5, 2, 8, 6, 1, 7, 9]
]
print("Solution valide" if est_valide(grille) else "Solution invalide")
```

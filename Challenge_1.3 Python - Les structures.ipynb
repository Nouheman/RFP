# 1. Calcul du montant d'une remise

Un commerçant accorde une remise de 5 % pour tout achat d’un montant compris entre 100 et 500 € et 8 % au-delà.<br>
Ecrivez un programme de calcul du montant de la remise sur un achat donné.<br>
Affichez la remise.

```python
#début , déclare la varible prixArticle.
prixArticle = int(input("Entrez le prix de l'article : "))
if(prixArticle>= 100 and prixArticle <= 500):
  print("Remise de 5% accordée soit : ",prixArticle*0.05,"€")
elif(prixArticle>500):
  print("Remise de 8% accordée soit : ",prixArticle*0.08,"€")
else:
  print("Aucune remise accordée")
#fin
```

# 4. Signe d'un produit

Ecrivez un programme qui affiche le signe du produit de deux nombres réels sans calculer la valeur de ce produit. Par signe, on entend positif, négatif ou nul.

```python
n = int(input("Entrez la valeur de n : "))
m = int(input("Entrez la valeur de m : "))
if(n*m>0):
  print("Le signe du produit est positif +")
elif(n*m<0):
  print("Le signe du produit est négatif -")
else:
  print("Le signe est nul donc l'un des entier est nul '0' ")
```

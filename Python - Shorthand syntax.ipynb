Une fonction qui convertit des montants en dollars vers des montants en euros (prends le cours euro/dollar du jour où tu fais la quête).
Une fonction qui renvoie la somme des nombres pairs présents dans une liste contenant des nombres entiers.
Une fonction qui convertit une liste de nombres entiers positifs ou négatifs) en une nouvelle liste contenant leurs valeurs absolues.
Tu peux prendre la liste suivante pour exemple :
test_list = [0, 3, -6, 9, 12, -15, 18, -21, 24, 27]

```python
def dollars_to_euros(amount):
    # Pour cette démonstration, on suppose que le taux de change est fixe à 1 USD = 0.88 EUR
    # Dans une application réelle, vous devriez récupérer le taux de change actuel
    exchange_rate = 0.88
    return amount * exchange_rate

print(dollars_to_euros(100))
```

```python
dollars_to_euros_short = lambda x: x * 0.88
print(dollars_to_euros_short(100))
```

```python
def sum_even_numbers(lst):
    sum = 0
    for num in lst:
        if num % 2 == 0:
            sum += num
    return sum

test_list = [0, 3, -6, 9, 12, -15, 18, -21, 24, 27]
print(sum_even_numbers(test_list))
```

```python
sum_even_numbers_short = lambda lst: sum(num for num in lst if num % 2 == 0)
test_list = [0, 3, -6, 9, 12, -15, 18, -21, 24, 27]
print(sum_even_numbers_short(test_list))
```

```python
def absolute_values(lst):
    return [abs(num) for num in lst]

test_list = [0, 3, -6, 9, 12, -15, 18, -21, 24, 27]
print(absolute_values(test_list))
```

```python
absolute_values_short = lambda lst: [abs(num) for num in lst]
test_list = [0, 3, -6, 9, 12, -15, 18, -21, 24, 27]
print(absolute_values_short(test_list))
```

# PIVOT TABLE

## It's like in Excel, but with Pandas

There are 2 syntaxes, choose your preferred :
- As a method : YOUR_DATAFRAME.pivot_table()
- As a function : pd.crosstab()

```
# Example
df_weather.pivot_table(values = 'MAX_TEMPERATURE_C'  , index = 'MONTH', aggfunc = 'mean')
```

## Create a Pivot table with :
- Months on rows
- the maximum of Max Temperature on values
- Opinion on columns

```
# Hop !
df_weather.pivot_table(index='MONTH', values='MAX_TEMPERATURE_C',  columns='OPINION')
```

## Pivot table with :
- Days on rows
- Precipitation on values
- Months on columns


```
# Hop !
df_weather.pivot_table(index='DAY', values='PRECIP_TOTAL_DAY_MM', columns='MONTH')
#error for precipitation
```

## Pivot table with :
- Months on rows
- the sum of Precipitations on values

**Then display it as a column chart.** (optionnal)

```
# on rajoute les import nécessaires sns et plt
import seaborn as sns
import matplotlib.pyplot as plt
# Hop !
```

```
# Hop !
df_weather.pivot_table(columns='MONTH', values='PRECIP_TOTAL_DAY_MM', aggfunc='sum')
```

## Pivot table with :
- Months on rows
- the average of Max Temperature on a first column of values
- the average of Sunhours on a second column of values

```
# Hop !
averageMAX_S = df_weather.pivot_table(index='MONTH', values=['MAX_TEMPERATURE_C', 'SUNHOUR'], aggfunc='mean')
print(averageMAX_S)
```

# Chart
Using your favorite library (MatPlotLib, Seaborn, Plotly or what you want), draw a chart from the previous pivot table with :
- Temperature in a blue column chart
- Sunhours in a red line plot

Tip : Once your pivot table created, you can help yourself with the `.reset_index()`  method.

```
# Your code here

mean_sunhour_maxtemp = averageMAX_S.reset_index()

sns.lineplot(data=mean_sunhour_maxtemp, x='MONTH', y='SUNHOUR', color='red')
sns.barplot(data=mean_sunhour_maxtemp, x='MONTH', y='MAX_TEMPERATURE_C', color='blue')

plt.title("Average Max Temperature with Average Sun Hours per Month")
plt.ylabel("Temperature (°C) , Sun by Hours")
plt.tight_layout()
```
